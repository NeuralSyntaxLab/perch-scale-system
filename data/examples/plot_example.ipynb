```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pathlib import Path

# Navigate to the exampe weight report CSV file
examples_dir = Path().resolve()
weight_report_path = examples_dir / "weight_report_example.csv"

# read the CSV file
df = pd.read_csv(weight_report_path)

# parse time column
df['Time'] = pd.to_datetime(df['Time'])

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['Time'], df['a_ro'], color='blue', alpha=0.6, s=5)
# Automatic date locator + custom formatter for hh:mm
locator = mdates.AutoDateLocator()
formatter = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)

fig.autofmt_xdate(rotation=45, ha='right')
ax.set_title(f"Example of a bird's weight data")
ax.set_xlabel('Time')
ax.set_ylabel('Weight (grams)')
fig.tight_layout()
plt.show()

# save figure to examples directory
fig.savefig(examples_dir / "weight_report_example_plot.png")
```
